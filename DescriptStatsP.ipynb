```python
#reading the excel file
#imported packages

import pandas as pd
import numpy as np

from scipy.stats.stats import pearsonr

from matplotlib import pyplot as plt

from mainP import Main

mn = Main()
file = mn.file
```

```python
#population growth for last 70 years

pg = pd.DataFrame(file, columns=['Population growth']).iloc[:71]
migr  = pd.DataFrame(file, columns=['Migration']).iloc[:71]
year = pd.DataFrame(file, columns=['year']).iloc[:71]


B_D  = pd.DataFrame(file).apply(lambda x: x['No Births All'] - x['All Deaths'], axis=1).iloc[:71]
plt.plot(year, pg, label='Population growth')
plt.plot(year, migr, label='Migration')
plt.plot(year, B_D, label='Births - Deaths')

plt.axvline(x=1998, color="black", linestyle=":")

plt.legend()
plt.show()
```

```python
#UK's population size for last 70 years

pop = pd.DataFrame(file, columns=['population']).iloc[:71]

year = pd.DataFrame(file, columns=['year']).iloc[:71]
plt.axvline(x=1998, color="black", linestyle=":")

plt.plot(year, pop)
plt.show()
```
