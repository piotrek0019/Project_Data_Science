```python
#reading the excel file
#imported packages

import pandas as pd
import numpy as np

from scipy.stats.stats import pearsonr

from matplotlib import pyplot as plt

from mainP import Main

mn = Main()
file = mn.file
```

```python
#population growth for last 70 years

pg = pd.DataFrame(file, columns=['Population_growth']).iloc[:71]
migr  = pd.DataFrame(file, columns=['Migration']).iloc[:71]
year = pd.DataFrame(file, columns=['year']).iloc[:71]


B_D  = pd.DataFrame(file).apply(lambda x: x['No_Births_All'] - x['All_Deaths'], axis=1).iloc[:71]
plt.plot(year, pg, label='Population_growth')
plt.plot(year, migr, label='Migration')
plt.plot(year, B_D, label='Births - Deaths')

plt.axvline(x=1998, color="black", linestyle=":")

plt.legend()
plt.show()
```

```python
#UK's population size for last 70 years

pop = pd.DataFrame(file, columns=['population']).iloc[:71]

year = pd.DataFrame(file, columns=['year']).iloc[:71]
plt.axvline(x=1998, color="black", linestyle=":")

plt.plot(year, pop)
plt.show()
```

```python
#central tendency for two periods

pb = pd.DataFrame(file, columns=['Population_growth']).iloc[1:22]
migr  = pd.DataFrame(file, columns=['Migration']).iloc[1:22]

print('Mean 1998 - 2018 for', pb.mean())
print('Median 1998 - 2018 for', pb.median())
print('Mean 1998 - 2018 for', migr.mean(), '\n')

pb = pd.DataFrame(file, columns=['Population_growth']).iloc[23:71]
migr  = pd.DataFrame(file, columns=['Migration']).iloc[23:71]
                                       
print('Mean 1949 - 1997 for', pb.mean())
print( 'Median 1949 - 1997 for', pb.median()) 
print('Mean 1949 - 1997 for', migr.mean())

```

```python
#Correlation of population growth and migration
pg = pd.DataFrame(file, columns=['Population_growth']).iloc[1:71]
migr  = pd.DataFrame(file, columns=['Migration']).iloc[1:71]

plt.scatter(pg, migr)

m,b = np.polyfit(np.hstack(pg.values), migr, 1)

plt.plot(pg, m*pg+b)
```

```python
#corellation of population growth and migration (matrix)
pg = pd.DataFrame(file, columns=['Population_growth','Migration']).iloc[1:71]

pg.corr()
```

```python
#corellation of net birth rate and migration

pg = pd.DataFrame(file, columns=['Population_growth']).iloc[1:71]
B_D  = pd.DataFrame(file).apply(lambda x: x['No_Births_All'] - x['All_Deaths'], axis=1).iloc[1:71]

plt.scatter(pg, B_D)

m,b = np.polyfit(np.hstack(pg.values), B_D, 1)

plt.plot(pg, m*pg+b)

```

```python
#corellation of net birth rate and migration (maxtrix)

pg = pd.DataFrame(file, columns=['Population_growth','Migration']).iloc[1:71]
pg['Births_minus_Deaths']  = pd.DataFrame(file).apply(lambda x: x['No_Births_All'] - x['All_Deaths'], axis=1).iloc[1:71]

pg.corr()
```
